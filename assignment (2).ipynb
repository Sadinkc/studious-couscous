```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL for Tesla's revenue data (adjust URL as needed)
url = 'https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue'

# Set a User-Agent header to mimic a browser
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
}
response = requests.get(url, headers=headers)

# Check if the request was successful
if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find the table containing the revenue data
    table = soup.find('table', {'class': 'historical_data_table'})

    if table:  # Check if the table was found
        # Extract table rows
        rows = table.find_all('tr')

        # Process each row to extract date and revenue
        revenue_data = []
        for row in rows[1:]:  # Skip header row
            columns = row.find_all('td')
            date = columns[0].text.strip()
            revenue = columns[1].text.strip().replace('$', '').replace(',', '')  # Clean revenue data
            revenue_data.append((date, float(revenue)))

        # Convert list to DataFrame for better handling
        tesla_revenue_df = pd.DataFrame(revenue_data, columns=['Date', 'Revenue'])
        tesla_revenue_df['Date'] = pd.to_datetime(tesla_revenue_df['Date'])

        print(tesla_revenue_df.head())
    else:
        print("Table not found on the page. Check if the website structure has changed.")
else:
    print(f"Failed to retrieve data. Status code: {response.status_code}")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL for GameStop's financials (adjust if needed)
url = 'https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue'

# Set a User-Agent header to mimic a browser
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
}
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, 'html.parser')

# Check if request was successful and print status code
print("Status Code:", response.status_code)

# Find and print the revenue data (this will depend on the page structure)
table = soup.find('table', {'class': 'historical_data_table'})

# Check if table was found
if table:
    rows = table.find_all('tr')

    revenue_data = []
    for row in rows[1:]:  # Skip the header row
        columns = row.find_all('td')
        date = columns[0].text.strip()
        revenue = columns[1].text.strip().replace('$', '').replace(',', '') # Clean up revenue string
        revenue_data.append((date, revenue))

    # Create the DataFrame here
    gamestop_revenue_df = pd.DataFrame(revenue_data, columns=['Date', 'Revenue'])
    gamestop_revenue_df['Revenue'] = pd.to_numeric(gamestop_revenue_df['Revenue']) # Convert Revenue to numeric
    gamestop_revenue_df['Date'] = pd.to_datetime(gamestop_revenue_df['Date']) # Convert Date to datetime

    print(gamestop_revenue_df.head())
else:
    print("Table not found. Check URL and website structure.")
```

```python
import matplotlib.pyplot as plt

# Check if the DataFrame is not empty
if not gamestop_revenue_df.empty:
    plt.figure(figsize=(12, 6))

    plt.plot(gamestop_revenue_df['Date'], gamestop_revenue_df['Revenue'], marker='o', linestyle='-', color='b')
    plt.title('GameStop Revenue Over Time')
    plt.xlabel('Date')
    plt.ylabel('Revenue (USD)')
    plt.grid(True)

    # Rotate date labels for better readability
    plt.xticks(rotation=45)
    plt.tight_layout()  # Adjust layout to prevent clipping of labels

    plt.show()
else:
    print("No data available to plot.")
```
